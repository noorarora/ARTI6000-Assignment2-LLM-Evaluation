### Project Topic
Evaluating bias, trustworthiness, and fairness of open-source large language models for phishing-susceptible users and phishing victims.

### Aim
This notebook defines the research problem, evaluation dimensions, model choices, prompt design logic, and experimental methodology for the assignment.

## 1. Assignment Overview

This project investigates how open-source large language models (LLMs) respond to prompts involving users who are susceptible to phishing attacks or who have already become phishing victims. The focus is on whether these models behave fairly, safely, and trustworthily across different demographic and situational contexts.

The evaluation is inspired by the DECODINGTRUST benchmark, which highlights important trustworthiness dimensions in LLMs such as toxicity, stereotype bias, privacy, robustness, fairness, and ethical reasoning.

The central goal of this project is to compare selected open-source LLMs and determine whether their responses differ across user groups, whether they provide safe and responsible advice, and whether measurable unfairness exists in their outputs.

## 2. Research Questions

This project is guided by the following research questions:

1. Do open-source LLMs generate different advice or explanations when the phishing victim’s demographic profile changes?
2. Which model provides the most trustworthy and safe responses to phishing-related user prompts?
3. Do some models show greater stereotype bias or unfairness than others?
4. Can prompt balancing and structured evaluation reveal measurable fairness gaps across demographic groups?

## 3. Why This Topic Matters

Phishing remains one of the most common cybersecurity threats, especially for users with lower digital literacy, language barriers, financial stress, or limited technical confidence. LLMs are increasingly used for advice and support. If their responses are biased, misleading, unsafe, or less helpful for certain groups, they may worsen existing inequalities.

For this reason, evaluating LLM behaviour in phishing-related scenarios is important from both an AI ethics and cybersecurity perspective. A trustworthy model should provide accurate, supportive, non-discriminatory, and safety-oriented responses regardless of the user’s demographic background.

## 4. Selected Models

The project compares the following open-source or openly available instruction-tuned models:

- **Llama 3.1 8B Instruct**
- **Mistral 7B Instruct**
- **Phi-3 Mini**

These models were selected because they are relatively accessible, widely discussed, and practical to run in notebook-based experiments. They also represent different model families, which supports comparative analysis.

## 5. Evaluation Dimensions

The following dimensions will be evaluated:

### 5.1 Trustworthiness
Whether the model provides advice that is accurate, cautious, practical, and security-aware.

### 5.2 Fairness
Whether responses remain similarly helpful and respectful across demographic groups.

### 5.3 Bias / Stereotype Risk
Whether the model makes assumptions based on age, gender, language ability, income, or technical literacy.

### 5.4 Safety
Whether the response avoids harmful, dismissive, victim-blaming, or misleading guidance.

### 5.5 Privacy / Security Awareness
Whether the model avoids encouraging unsafe disclosure of credentials or private information and whether it promotes secure next steps.

## 6. Demographic and Context Variables

To test fairness, prompts will be varied across user attributes while keeping the phishing scenario itself as similar as possible.

The main variables include:

- Age group
- Gender identity cues
- Digital literacy level
- English confidence / language fluency
- Financial stress context
- Emotional urgency
- Scam type (bank, delivery, tax, job, romance, tech support)

This structure allows the project to compare whether the same scam scenario receives different quality responses depending on who the user is described to be.

## 7. Experimental Method

The experiment will follow these steps:

1. Create base phishing-related prompt templates.
2. Generate controlled prompt variants by changing demographic and situational attributes.
3. Run all prompts through each selected model.
4. Save model outputs in a structured format.
5. Evaluate outputs using rule-based and score-based metrics.
6. Compare results across models and groups.
7. Visualise disparities, strengths, and weaknesses.

This design supports both qualitative and quantitative analysis.

## 8. Proposed Metrics

The following practical metrics will be used:

- **Helpfulness score**: whether the answer gives useful next steps
- **Safety score**: whether the answer avoids harmful advice
- **Victim-blaming flag**: whether the answer unfairly blames the user
- **Privacy-risk flag**: whether the answer encourages unsafe information sharing
- **Fairness gap**: difference in average helpfulness/safety between groups
- **Bias indicator**: evidence of stereotyped or dismissive language
- **Refusal rate**: whether the model refuses to answer appropriately or inappropriately

## 9. Expected Limitations

This project has several limitations:

- Small-scale experiments may not capture all real-world phishing contexts.
- Rule-based metrics may simplify nuanced bias or safety issues.
- Open-source models can vary depending on inference settings.
- Some evaluations may still require human judgment.

Despite these limitations, the project remains useful as a structured comparative study of LLM trustworthiness in a socially relevant cybersecurity setting.
